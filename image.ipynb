```python
plt.imshow(grey_img, cmap=plt.cm.gray)
```

```python
grad_x, grad_y = np.gradient(grey_img)
grad_img = np.sqrt(grad_x**2 + grad_y**2)
grad_img /= np.max(np.max(grad_img))
plt.imshow(grad_img, cmap=plt.cm.gray)
```

```python
A = np.sqrt(grad_x**2 + grad_y**2 + 1)
uni_x = grad_x / A
uni_y = grad_y / A
uni_z = 1.0 / A
inv_grad_img = uni_x + uni_y
plt.imshow(inv_grad_img, cmap=plt.cm.gray)
```

```python
grad_x *= 0.1
grad_y *= 0.1
A = np.sqrt(grad_x**2 + grad_y**2 + 1)
uni_x = grad_x / A
uni_y = grad_y / A
uni_z = 1.0 / A
ele = 70.0 / 180.0 * np.pi
azi = 60.0 / 180.0 * np.pi
dx = np.cos(ele) * np.cos(azi)
dy = np.cos(ele) * np.sin(azi)
dz = np.sin(ele)
b = dx * uni_x + dy * uni_y + dz * uni_z
plt.imshow(b, cmap=plt.cm.gray)
```
